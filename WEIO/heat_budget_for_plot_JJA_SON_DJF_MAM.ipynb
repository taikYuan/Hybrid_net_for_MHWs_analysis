```python
import numpy as np

# 假设 data1 和 data2 是两个二维数组，且它们具有相同的行数
# 这里我们使用随机数据来模拟这两个数组
# np.random.seed(0)  # 设置随机种子以便结果可复现
data1 = Atlantic_DJF[:26] # 100个样本，每个样本1个特征
data2 = low_cloud_cover  # 与data1相同数量的样本

# 标准化数据
z1 = (data1 - np.mean(data1, axis=0)) / np.std(data1, axis=0, ddof=1)
z2 = (data2 - np.mean(data2, axis=0)) / np.std(data2, axis=0, ddof=1)

# 计算协方差矩阵
cov_matrix = np.cov(z1, z2, rowvar=False)

# 提取两个变量之间的协方差
covariance = cov_matrix[0, 1]

# 计算标准差
std1 = np.std(z1, ddof=1)
std2 = np.std(z2, ddof=1)

# 计算相关系数
corr = covariance / (std1 * std2)

print("协方差:", covariance)
print("标准差 data1:", std1)
print("标准差 data2:", std2)
print("Pearson相关系数为:", corr)
```
